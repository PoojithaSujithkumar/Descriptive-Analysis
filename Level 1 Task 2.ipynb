```python
df.describe()
```

```python
print("Median values:\n", df.median(numeric_only=True))
print("\nStandard deviation:\n", df.std(numeric_only=True))
```

```python
for col in ['Country Code', 'City', 'Cuisines']:
    if col in df.columns:
        print(f"\nUnique values and frequency in '{col}':\n")
        print(df[col].value_counts())
    else:
        print(f"'{col}' column not found in dataset.")
```

```python
if 'City' in df.columns:
    top_cities = df['City'].value_counts().head(10)
    print("\nTop 10 Cities with Most Restaurants:\n", top_cities)

    top_cities.plot(kind='bar', figsize=(10,5), title='Top Cities')
else:
    print("City column not found.")
```

```python
if 'Cuisines' in df.columns:
    top_cuisines = df['Cuisines'].value_counts().head(10)
    print("\nTop 10 Cuisines:\n", top_cuisines)

    # Bar plot
    top_cuisines.plot(kind='bar', figsize=(10,5), color='orange', title='Top Cuisines')
else:
    print("Cuisines column not found.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

if 'Country Code' in df.columns:
    plt.figure(figsize=(8,4))
    sns.countplot(data=df, x='Country Code')
    plt.title('Country Code Distribution')
    plt.show()
```
